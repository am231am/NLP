# Tokenization in Natural Language Processing (NLP)  

**Tokenization** is the process of breaking text into smaller units called **tokens**, which can be words, phrases, or individual characters. This step is crucial for enabling machines to understand and process human language effectively.  

## ðŸ“Œ Types of Tokenization  

### ðŸ”¹ Word Tokenization  
Splits text into individual words.  

**Example:**  
> *Original Sentence:*  
> `"The quick brown fox jumps over the lazy dog"`  
> *Tokenized Output:*  
> `["The", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog"]`  

---  

### ðŸ”¹ Sentence Tokenization  
Divides text into sentences.  

**Example:**  
> *Original Text:*  
> `"Hello world. How are you?"`  
> *Tokenized Output:*  
> `["Hello world.", "How are you?"]`  

---  

### ðŸ”¹ Subword Tokenization  
Breaks down words into **subword units**, which is useful for handling complex words or out-of-vocabulary terms.  
Techniques like **Byte Pair Encoding (BPE)** are commonly used for this purpose.  

**Example:**  
> *Original Word:* `"unhappiness"`  
> *Tokenized Output:* `["un", "happi", "ness"]`  

---  

### ðŸ”¹ Character Tokenization  
Splits text into **individual characters**, treating each as a separate token. This is particularly useful for:  
- **Languages without clear word boundaries (e.g., Chinese, Japanese, Thai)**

# [**Stopword Removal in NLP**](https://www.analyticsvidhya.com/blog/2019/08/how-to-remove-stopwords-text-normalization-nltk-spacy-gensim-python/#h-what-are-stopwords)

**stopwords** are common words that are often removed from text before processing because they usually do not carry significant meaning. These words include articles, prepositions, conjunctions, and pronouns like "the," "is," "in," "and," "of," "to," etc.

**Why Remove Stopwords?**
# Why Remove Stopwords?

## 1. Reduces Noise  
Stopwords donâ€™t add much meaning to a sentence. Removing them makes the sentence clearer.

**Example:**  
- **Before:** *"The cat is sitting on the table."*  
- **After Removing Stopwords:** *"cat sitting table."*  

The main words **"cat sitting table"** still keep the meaning, but the unnecessary words (stopwords) like **"the," "is," "on"** are removed.

---

## 2. Improves Efficiency  
When we remove stopwords, the system processes **fewer words**, making text analysis faster.

Imagine we have a chatbot processing **1 million sentences** per day.  

### **Example:**  
- **Without Removing Stopwords:**  
   "I want to know the weather in New York today." â†’ **10 words**  
- **With Stopwords Removed:**  
   "want know weather New York today." â†’ **5 words**  

By processing **half the words**, the chatbot works **twice as fast** and needs **less memory**.

---

## 3. Enhances Model Performance  
Stopwords can confuse NLP models. Removing them makes models more accurate.  

### **Example: Sentiment Analysis (Good vs. Bad Reviews)**  
- **Without Removing Stopwords:**  
   "I am not happy with this product."  
  The model might focus on **"happy"** and wrongly classify it as **positive**.

- **With Stopwords Removed:**  
   "not happy product."  
  Now, it clearly shows a **negative** meaning.

---


### [**Stopword Removal using NLTK**](https://www.analyticsvidhya.com/blog/2019/08/how-to-remove-stopwords-text-normalization-nltk-spacy-gensim-python/#h-what-are-stopwords)

NLTK has a list of stopwords stored in 16 different languages.

````python
# importing libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
````

````python
# Download NLTK resources (only needed once)
nltk.download('punkt_tab')
nltk.download('stopwords')
````

````python
print(stopwords.words('english'))
print(stopwords.words('arabic'))
````

````python
# Example sentences
# sample sentence
text = """He determined to drop his litigation with the monastry, and relinguish his claims to the wood-cuting and
fishery rihgts at once. He was the more ready to do this becuase the rights had become much less valuable, and he had
indeed the vaguest idea where the wood and river in question were."""
````

````python
# Tokenize the sentences
tokens = word_tokenize(text)
print("Before stopword removal:")
print("Sentence :", tokens)
````

````python
# Remove stopwords
stop_words = set(stopwords.words("english")) - {"not"}

filtered = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter stopword removal:")
print("Sentence :", filtered)
````

````python
print("Filtered Sentence: ")
print(" ".join(filtered))
````

# **[Stemming](https://www.analyticsvidhya.com/blog/2022/06/stemming-vs-lemmatization-in-nlp-must-know-differences/)**

is a fundamental technique in Natural Language Processing (NLP) that reduces words to their base or root form, known as the "stem." This process involves removing prefixes and suffixes (affixes) from words to treat different grammatical forms of a word as a single entity.



*   Prefix: A set of letters added to the beginning of a word to change its
    meaning.

    Example: Adding "un-" to "happy" makes "unhappy," meaning not happy.
*   Suffix: A set of letters added to the end of a word to change its form or
    meaning.
    Example: Adding "-ful" to "joy" makes "joyful," meaning full of joy.

***Examples:***

- "Running," "runner," and "ran" all come from the root word "run."

- "Happily" and "happiness" both come from "happy."



### **Stemming using nltk**

````python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
````

````python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

paragraph = """I have three visions for India. In 3000 years of our history, people from all over  the world have come and invaded us, captured our lands, conquered our minds.
From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British, the French, the Dutch, all of them came and looted us, took over what was ours.
Yet we have not done this to any other nation. We have not conquered anyone. We have not grabbed their land, their culture, their history and tried to enforce our way of life on them.  """

tokens = word_tokenize(paragraph)
print(tokens)
````

````python
filtered = [word for word in tokens if word.lower() not in stop_words]
print(filtered)
````

````python
stemmed_words = [stemmer.stem(word) for word in filtered]
print(' '.join(stemmed_words))
````

using SnowballStemmer

````python
from nltk.stem.snowball import SnowballStemmer

print(SnowballStemmer.languages)
````

````python
# Define a sample text
text = "running jumped happily swimming"

# Initialize the stemmer for English
stemmer = SnowballStemmer("english")

# Apply stemming word by word
stemmed_words = [stemmer.stem(word) for word in text.split()]
print("Original Text:", text)
print("Stemmed Words:", " ".join(stemmed_words))
````
