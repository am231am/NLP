## **First Step in an NLP Application**
### **1. Text Preprocessing (Data Cleaning)**
Before applying any NLP techniques, the raw text must be cleaned and structured to improve model accuracy and efficiency. This step ensures that the text is in a suitable format for further processing.

### **Key Steps in Text Preprocessing:**
1. **Text Cleaning (Using Regex)**  Remove special characters, punctuations (. , ! $( ) * % @), and URLs.
2. **Lowercasing**  Convert all text to lowercase for consistency.
3. **Tokenization**  Split text into individual words or sentences.
4. **Text Cleaning**  Remove special characters, punctuation, and unnecessary whitespace.
5. **Stopword Removal**  Eliminate common words like *the, is, and* that do not add significant meaning.
6. **Stemming & Lemmatization**  Reduce words to their root forms (e.g., *running*  *run*).

**Basic patterns that match single chars**


| Character  | function |
| ------------- | ------------- |
| a-z, 0-9  | ordinary characters just match themselves exactly.|
| . (dot)  | matches any single character except newline '\n'  |
| \w | matches a "word" character: a letter or digit or underbar [a-zA-Z0-9_] |
| \W | matches any non-word character |
| \b | boundary between word and non-word |
| \s | matches a single whitespace character -- space, newline, return, tab |
| \S | matches any non-whitespace character |
| \t, \n, \r | tab, newline, return |
| \d | decimal digit [0-9] |
| ^ | matches start of the string |
| $ | match the end of the string |

### import `re` module

````python
import re
````

````python
# ordinary characters just match themselves exactly
text = "hello world 1 23 world"
match = re.findall(r'\d+', text)
print(match)
````

````python
# Dot . (Matches Any Character Except Newline)
text = "cat,cft ,bat, hat, rat"
match = re.findall(r'c.t', text)
print(match)  # Output: ['cat']
````

````python
#\w (Matches Any Word Character [a-zA-Z0-9_])
text = "Hello_World 123!"
match = re.findall(r'\w+', text)
print(match)
````

````python
#\s (Matches Whitespace Characters)
text = "Hello World Python"
match = re.findall(r'\s', text)
print(match)
````

````python
#\S (Matches Any Non-Whitespace Character)
text = "H ello World !"
match = re.findall(r'\S+', text)
print(match)
````

````python
#^ (Matches Start of String)
text = "Python is a programming language"
match = re.findall(r'^P', text)
print(match)
````

**The search() Function**

- The search() function searches the string for a match, and returns a Match object if there is a match.

- If there is more than one match, only the first occurrence of the match will be returned:

````python
text = "I love  regex and i love Python  "
result_search = re.search(r'love', text)
print(result_search)
print(result_search.group())
````

The (\\.) in regex is used to match a literal dot (.) because in regex, a plain dot (.) is a special wildcard character that matches any single character (except a newline). By escaping it with a backslash (\.), we ensure it matches only an actual period (.).

````python
text = 'My mail is amina123@gmail.com and mohamed52@gmail.com'

print(re.search(r'(\w+)@(\w+)\.(\w+)', text))
print(re.search(r'\w+@\w+\.\w+', text).group())
````

````python
print(re.search(r'(\w+)@(\w+)\.(\w+)', text).group(0))
print(re.search(r'(\w+)@(\w+)\.(\w+)', text).group(1))
print(re.search(r'(\w+)@(\w+)\.(\w+)', text).group(2))
print(re.search(r'(\w+)@(\w+)\.(\w+)', text).group(3))
print(re.search(r'(\w+)@(\w+)\.(\w+)', text).group(1, 3))
````

## Tokenization Using Regular Expressions(RegEx) <a id ="5"></a>



### Word Tokenization

````python
text = """There are multiple ways we can perform tokenization on given text data.
We can choose any method based on langauge, library and purpose of modeling."""
tokens = re.findall("\w+", text)
print(tokens)
````

Based on RegEx pattern we are able to generate the list of words. Details about each character in our RegEx pattern is as below.
```
\w :    Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character).
+  :	One or more occurrences.
```

So our RegEx pattern signifies that the code should find all the alphanumeric characters until any other character is encountered.

### Sentence Tokenization

````python
text = """Characters like periods,
exclamation point and newline char are
 used to separate the sentences. But one drawback with split() method, that we can only use one separator at a time! So sentence tonenization wont be foolproof with split() method."""
# Pattern Explanation: [.!?]  Matches any one of the characters . ! ?. ' '  Ensures that the punctuation is followed by a space, preventing incorrect splits (e.g., "U.S.A.").
tokens_sent = re.compile('[.!?] ') # Using compile method to combine RegEx patterns
tokens_sent.split(text)
````

As you can see from above result, we are able to split sentence using multiple separators.
